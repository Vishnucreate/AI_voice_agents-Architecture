```python
import nemo.collections.asr as nemo_asr

asr_model = nemo_asr.models.EncDecCTCModel.from_pretrained(model_name="QuartzNet15x5Base-En")
```

```python
import srt
import json

with open("football.en.srt", "r") as f:
    subs = list(srt.parse(f.read()))

with open("manifest.json", "w") as out:
    for sub in subs:
        start = sub.start.total_seconds()
        end = sub.end.total_seconds()
        duration = round(end - start, 2)

        if duration > 0.3:  # skip too short lines
            out.write(json.dumps({
                "audio_filepath": "football.wav",
                "offset": round(start, 2),
                "duration": duration,
                "text": sub.content.replace("\n", " ")
            }) + "\n")
```

```python
from omegaconf import OmegaConf

config = OmegaConf.create({
    "model": {
        "train_ds": {
            "manifest_filepath": "manifest.json",
            "sample_rate": 16000,
            "labels": list(" abcdefghijklmnopqrstuvwxyz'"),
            "batch_size": 8,
            "shuffle": True,
            "num_workers": 2
        },
        "validation_ds": {
            "manifest_filepath": "manifest.json",
            "sample_rate": 16000,
            "labels": list(" abcdefghijklmnopqrstuvwxyz'"),
            "batch_size": 8,
            "shuffle": False,
            "num_workers": 2
        },
        "optim": {
            "lr": 0.001
        }
    },
    "trainer": {
        "max_epochs": 5,
        "gpus": 1
    }
})
```

```python
asr_model.setup_training_data(train_data_config=config.model.train_ds)
asr_model.setup_validation_data(val_data_config=config.model.validation_ds)
```

```python
asr_model.train()

```

```python
asr_model.transcribe(["football.wav"])
```
